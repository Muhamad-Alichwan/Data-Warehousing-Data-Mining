```python
#import library
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv("iris.csv")
df.head()
```

```python
df = df.drop(columns = "Id", axis = 1)
```

```python
df.head()
```

```python
sns.relplot(data = df, x = "SepalLengthCm", y = "SepalWidthCm", hue="Species")
```

```python
sns.heatmap(data = df.corr(), annot = True)
```

```python
#menentukan X dan y
X = df.drop(columns = "Species", axis = 1)
y = df["Species"]
```

```python
#encoding y
le = LabelEncoder()
y = le.fit_transform(y)
y
```

```python
#proses pembagian data test dan data train
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2, random_state = 0)
```

```python
#membuat model data
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
#Hasil prediksi dari data test dan data train
pred_train = model.predict(X_train)
pred_test = model.predict(X_test)

print(f"Hasil dari data train\n{classification_report(y_train, pred_train)}")
print(f"Hasil dari data test\n{classification_report(y_test, pred_test)}")
```
