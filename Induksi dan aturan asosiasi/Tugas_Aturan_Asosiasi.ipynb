```python
#Import Library
import numpy as np
import pandas as pd
from apyori import apriori
```

```python
df=pd.read_csv('GroceryStoreDataSet.csv', names = ["product"], header=None)
df
```

```python
records = list(df['product'].apply(lambda x: x.split(',')))
print(records)
```

```python
association_rules_1 = apriori(records, min_support= 0.003, min_confidence= 0.2, min_lift=3, min_length=2)
association_rules_2 = apriori(records, min_support= 0.007, min_confidence= 0.6, min_lift=3, min_length=2)

association_results_1 = list(association_rules_1)
association_results_2 = list(association_rules_2)
```

```python
print("association_results_1")
print(f"{association_results_1[0]}\n")
print(f"{association_results_1[1]}\n")
print(f"{association_results_1[2]}\n")
print(f"{association_results_1[3]}\n")
print(f"{association_results_1[4]}\n")

print("association_results_2")
print(f"{association_results_2[0]}\n")
print(f"{association_results_2[1]}\n")
print(f"{association_results_2[2]}\n")
print(f"{association_results_2[3]}\n")
print(f"{association_results_2[4]}\n")
```

```python
for item in association_results_1:
    pair = item[0]
    items = [x for x in pair]

    print(f"Rule: {items[0]} -> {items[1]}")
    print(f"Support: {str(item[1])}")
    print(f"Confidence: {str(item[2][0][2])}")
    print(f"Lift: {str(item[2][0][3])}")
    print("================================================================================")
```

```python
for item in association_results_2:
    pair = item[0]
    items = [x for x in pair]

    print(f"Rule: {items[0]} -> {items[1]}")
    print(f"Support: {str(item[1])}")
    print(f"Confidence: {str(item[2][0][2])}")
    print(f"Lift: {str(item[2][0][3])}")
    print("================================================================================")
```
