```python
#import library
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("Mall_Customers.csv")
df.head()
```

```python
df = df.filter(["Annual Income (k$)","Spending Score (1-100)"], axis = 1)
df.head()
```

```python
sns.scatterplot(data = df, x="Annual Income (k$)",y="Spending Score (1-100)" )
```

```python
#KMeans use elbow method
wcss = []
jarak = range(1,11)

for i in jarak:
    model = KMeans(n_clusters = i)
    model.fit(df)
    wcss.append(model.inertia_)

plt.plot(jarak, wcss)
plt.ylabel("wcss")
plt.xlabel("Number of clusters")
plt.show()
```

```python
#make model
model = KMeans(n_clusters = 5)
model.fit(df)
```

```python
#make scatterplot use model has been created
sns.scatterplot(data = df, x="Annual Income (k$)",y="Spending Score (1-100)" , hue = model.labels_, palette = "rainbow")
sns.scatterplot(x=model.cluster_centers_[:,0],y=model.cluster_centers_[:,1], c = "black")
```
